```python
import matplotlib.pyplot as plt

# Accuracy data
accuracy1 = [0.4719, 0.5589, 0.7236, 0.7574, 0.8414]
accuracy2 = [0.4708, 0.5619, 0.7185, 0.7408, 0.8280]
accuracy3 = [0.4708, 0.5803, 0.7400, 0.7615, 0.8465]
accuracy4 = [0.5824, 0.6407, 0.7421, 0.7697, 0.8557]
accuracy5 = [0.5947, 0.6479, 0.7584, 0.7738, 0.8618]

x_values = [10, 30, 100, 200, 350]
x_labels = ['10', '100', '500', '1000', '7911']
plt.plot(x_values, accuracy1, color='red', marker='o', label='bert', linewidth=1)
plt.plot(x_values, accuracy2, color='blue', marker='o', label='distill', linewidth=1)
plt.plot(x_values, accuracy3, color='green', marker='o', label='roberta', linewidth=1)
plt.plot(x_values, accuracy4, color='orange', marker='o', label='biobert', linewidth=1)
plt.plot(x_values, accuracy5, color='purple', marker='o', label='clinicalbert', linewidth=1)

# 添加两条基线线
plt.axhline(y=0.73, linestyle='--', color='grey', label='zero shot ChatGpT', linewidth=1)
plt.axhline(y=0.78, linestyle='--', color='black', label='zero shot GPT-4', linewidth=1)

# 添加图例
# plt.legend(loc='upper left')
plt.legend(loc='lower right')

# x轴参数
plt.xticks(x_values, x_labels)

# 设置标题和标签
plt.title('Comparsion between GPT and fine-tuned models')
plt.xlabel('# of sample in training')
plt.ylabel('Accuracy')

# 显示图形
plt.show()
```
